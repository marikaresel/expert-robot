```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('/Users/marika/World Health Organization Analysis/Exported Data/df_crude_rates_merged.csv')

# Identify and exclude non-numeric columns
numeric_columns = df.select_dtypes(include=['number'])

# Compute the correlation matrix
correlation_matrix = numeric_columns.corr()

# Create heatmap with annotations
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', annot_kws={"size": 10}, linewidths=0.5, linecolor='black')
plt.title('Correlation Matrix Heatmap')
plt.show()

# Ensure the directory exists
path = '/Users/marika/World Health Organization Analysis/Visuals'
if not os.path.exists(path):
    os.makedirs(path)

# Save the plot to the "Visuals" folder
file_name = os.path.join(path, 'correlation_matrix_heatmap.png')
plt.savefig(file_name)
```

```python
# The correlation between all the deaths are high
```

```python
# Find variables with strongest correlations
strongest_correlations = correlation_matrix.unstack().sort_values(ascending=False).drop_duplicates()
```

```python
# Select variables with strongest positive correlation
strongest_positive = strongest_correlations[(strongest_correlations < 1) & (strongest_correlations > 0)]
if not strongest_positive.empty:
    strongest_positive = strongest_positive.index[0]
else:
    strongest_positive = None
```

```python
# Select variables with strongest negative correlation
strongest_negative = strongest_correlations[(strongest_correlations < 0)]
if not strongest_negative.empty:
    strongest_negative = strongest_negative.index[0]
else:
    strongest_negative = None
```

```python
# Plot scatterplots for the strongest correlations if they exist
plt.figure(figsize=(12, 5))
```

```python
if strongest_positive:
    plt.subplot(1, 2, 1)
    sns.scatterplot(x=df_merged[strongest_positive[0]], y=df_merged[strongest_positive[1]])
    plt.title(f'Scatterplot of {strongest_positive[0]} vs {strongest_positive[1]}')
```

```python
if strongest_negative:
    plt.subplot(1, 2, 2)
    sns.scatterplot(x=df_merged[strongest_negative[0]], y=df_merged[strongest_negative[1]])
    plt.title(f'Scatterplot of {strongest_negative[0]} vs {strongest_negative[1]}')
```

```python
plt.tight_layout()
plt.show()
```

```python
# Seems to be an upward trend 
```

```python
# Categorize variable A
bins = [0, 2, 4, 6]
labels = ['Low', 'Medium', 'High']
df_merged['A_category'] = pd.cut(df_merged['Value_10'], bins=bins, labels=labels, right=False)
```

```python
# Create a categorical plot
sns.catplot(x='A_category', y='FactValueNumericLow', kind='bar', data=df_merged)
plt.title('Categorical Plot of A_category vs B')
plt.xlabel('Category of A')
plt.ylabel('Value of B')
plt.show()
```

```python
# Low Category: When A is in the Low range, the average value of FactValueNumericLow is 10.
Medium Category: When A is in the Medium range, the average value of FactValueNumericLow increases to 20.
High Category: When A is in the High range, the average value of FactValueNumericLow further increases to 30.
```
